## 1. What is `QSDsan`
`QSDsan` is an open-source, community-led platform for the quantitative sustainable design (QSD) of sanitation and resource recovery systems.

You can find:
- Source codes of `QSDsan` on its [GitHub repository](http://github.com/qsd-group/qsdsan)
- Documentation of `QSDsan` on [readthedocs](https://qsdsan.readthedocs.io)

## 3. Installation

## 4. A simple example

To test if you have successfully installed ``QSDsan``, in your IDE, do:

```python
import qsdsan as qs
qs.__version__ # this tells you which version of QSDsan you installed
```

The library [``EXPOsan``](https://github.com/QSD-group/exposan) has sanitation systems developed using ``QSDsan``, you can also check out those systems (make sure you ``pip install exposan`` first), e.g.,

```python
# Three sanitation alternatives as described in Trimmer et al., 
# Environ. Sci. Technol. 2020, 54 (19), 1264112653.
# https://doi.org/10.1021/acs.est.0c03296.
from exposan import bwaise as bw
bw.sysA.diagram('cluster')
```

```python
bw.print_summaries((bw.sysA, bw.sysB, bw.sysC))
```

If you can see the system diagram and printed summaries, then congrats! You've successfully installed ``QSDsan`` and continue on the modules!
