```python
# Import libraries
import numpy as np, pandas as pd
from matplotlib import pyplot as plt
```

```python
# Import data and do a bit cleanup
data = pd.read_excel('MCDA_Solution.xlsm', sheet_name='Macro', header=1, skiprows=0, usecols=np.arange(1, 10, 1))
names = list(data.columns)
names[-1] = 'Winner'
data.columns = names
```

```python
# Function to plot the criterion weight combinations where a certain alternative wins
def plot_winner(alt='A', color='k', linewidth='0.5', **kwargs):
    fig, ax = plt.subplots()
    alt = alt.upper()
    winning_wts = data[data['Winner']==alt].iloc[:, :5].values
    for winning_wt in winning_wts:
        ax.plot(winning_wt, color='k', linewidth='0.5', **kwargs)
    ax.set_xticks(np.arange(0, 5, 1))
    ax.set_xticklabels(names[:5])
    ax.set_ylabel('Criterion Weight')
    ax.set_title(f'Winning Weights for Alternative {alt}')
    return fig, ax
```

```python
figA, axA = plot_winner('A')
```

```python
figB, axB = plot_winner('B')
```

```python
fig, axC = plot_winner('C')
```
