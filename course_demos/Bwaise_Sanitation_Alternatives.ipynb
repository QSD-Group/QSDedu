# Sanitation Alternatives in Bwaise <a class="anchor" id="top"></a>

This module is a simple demonstration of the three sanitation alternative systems as described in [Trimmer et al.](https://doi.org/10.1021/acs.est.0c03296) All source codes are in the [`bwaise`](https://github.com/QSD-Group/EXPOsan/tree/main/exposan/bwaise) module in the EXPOsan package.

```python
import qsdsan as qs, exposan as es
print(f'This tutorial was made with qsdsan v{qs.__version__}, exposan v{es.__version__}.')
```

```python
# Load Bwaise systems
from exposan import bwaise as bw
bw.load()
```

```python
# Retrieving the different systems
sysA, sysB, sysC = bw.sysA, bw.sysB, bw.sysC
```

```python
# Quick summary of each system
bw.print_summaries((sysA, sysB, sysC))
```

```python
# Flow diagram, this might not show due to graphviz installation issues
sysA.diagram()
```

```python
# Take a look at each unit
bw.A2.show()
```

```python
# Cost
bw.A2.results()
```

```python
teaA = sysA.TEA
teaA.show()
teaA.annualized_NPV
```

```python
lcaA = sysA.LCA
lcaA.show()
```

```python
lcaA.get_impact_table('construction')
```
